```python
import matplotlib.pyplot as plt
```

```python
x=[23,34,22,53,35,5,98,32,86,26]
y=[7,74,43,26,87,22,24,79,34,46]

plt.plot(x, marker='*',ms=10 , mec='red', mfc='y',label='Plot for value of X')
plt.plot(y, marker='o',ms=10, mec='blue', mfc='k', label='Plot for values of Y')
```

```python
plt.scatter(x,y,color='orange')
```

```python
plt.bar(x,y, color='black', width=20)
```

```python
plt.barh(y,x, color='red', width=30)
```

```python
plt.pie(x)
```

```python
plt.pie(x, label='Pie chart')
```
